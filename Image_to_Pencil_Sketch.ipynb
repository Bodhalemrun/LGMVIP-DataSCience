```python
import cv2
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
import requests
print('Start Downloading')
# url = 'https://images.unsplash.com/photo-1529778873920-4da4926a72c2?ixlib=rb-1.2.1&ixid=MnwxMjA3fDB8MHxwaG90by1wYWdlfHx8fGVufDB8fHx8&auto=format&fit=crop&w=436&q=80'
url = 'https://d17fnq9dkz9hgj.cloudfront.net/uploads/2020/04/shelter-dog-cropped-1-768x400.jpg'
r = requests.get(url)
filename = 'image.jpg'
with open(filename, 'wb') as out_file:
    out_file.write(r.content)
    
print('Download Completed')
```

```python
image = cv2.imread(filename)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
#plt.figure(figsize = (8,8))
plt.imshow(image)
plt.axis("off")
plt.title("Original Image")
plt.show()
# cv2.imshow("Original Image", image)
# cv2.waitKey(0)
```

**Convert the Image into Grayscale**

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.figure(figsize = (8,8))
plt.imshow(gray_image, cmap = "gray")
plt.axis("off")
plt.title("GrayScale Image")
plt.show()
```

**Inverting the graysacle Image**

```python
inverted_image = cv2.bitwise_not(gray_image)
plt.figure(figsize = (8,8))
plt.imshow(inverted_image, cmap = "gray")
plt.axis("off")
plt.title("Inverted Image")
plt.show()
```

**Blurring the Inverted Image**

```python
image_smoothing = cv2.GaussianBlur(inverted_image, (21,21),sigmaX = 0, sigmaY = 0)
plt.figure(figsize = (8,8))
plt.imshow(image_smoothing, cmap = "gray")
plt.axis("off")
plt.title("Image_Smoothing")
plt.show()
```

**Creating Pencil Sketch**

```python
pencil_sketch = cv2.divide(gray_image, 255 - image_smoothing ,scale = 255)
plt.figure(figsize = (8,8))
plt.imshow(pencil_sketch, cmap = "gray")
plt.axis("off")
plt.title("Final Pencil Sketch")
plt.show()
```

**Final Result**

```python
plt.figure(figsize = (50,50))
plt.subplot(1,5,1)
plt.imshow(image)
plt.axis("off")
plt.title("Original Image")
plt.subplot(1,5,2)
plt.imshow(gray_image, cmap = "gray")
plt.axis("off")
plt.title("GrayScale Image")
plt.subplot(1,5,3)
plt.imshow(inverted_image, cmap = "gray")
plt.axis("off")
plt.title("Inverted Image")
plt.subplot(1,5,4)
plt.imshow(image_smoothing, cmap = "gray")
plt.axis("off")
plt.title("Image_Smoothing")
plt.subplot(1,5,5)
plt.imshow(pencil_sketch, cmap = "gray")
plt.axis("off")
plt.title("Final Pencil Sketch")
plt.show()
```
